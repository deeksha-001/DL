```python
import tensorflow as tf
from tensorflow import keras #API for building and training DL models
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
# Converts text to numbers
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences 
#LSTM models require fixed length input sequences, so we need to pad our sequences to ensure they all have the same length.
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Download the dataset

path = tf.keras.utils.get_file(
    'shakespeare.txt',
    'https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt'
)
```

```python
text = open(path,'r',encoding = 'utf-8').read()
```

```python
# Convert text into lowercase

text = text.lower()

# Display the first 500 characters

print(text[:500])
```

```python
print("Total characters:", len(text))
```

```python
print("Total words:", len(text.split()))
```

```python
tokenizer = Tokenizer()

tokenizer.fit_on_texts([text])
```

```python
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1

print("Total unique words:", vocab_size)
```

```python
for word,index in list(word_index.items())[:10]:
    print(word,":",index)
```

```python
# Convert text into sequences

sequences = tokenizer.texts_to_sequences([text])[0]

print(sequences[:30])
```

```python
# Split text into lines

lines = text.split('\n')

# Convert lines into sequences

input_sequences = []

for line in lines:
    # Convert line into sequence
    token_list = tokenizer.texts_to_sequences([line])[0]

    # Create n-grams
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence) 

print("total input sequences:", len(input_sequences)) 
```

Part 2 

#### Padding -> Train/Test -> LSTM Model -> Training

```python
# Find the max sequence length

max_sequence_len = max([len(x) for x in input_sequences])
print("Max sequence length:", max_sequence_len)
```

```python
# Apply padding

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
)

print("sequences and words per sequence:", input_sequences.shape)
```

```python
#  Split into x & y

x = input_sequences[:,:-1]

y = input_sequences[:,-1]

print("x shape:", x.shape)
print("y shape:", y.shape)
```

```python
y = tf.keras.utils.to_categorical(y, num_classes=vocab_size)

print(y.shape)
```
